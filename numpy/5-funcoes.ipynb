```python
import numpy as np
```

```python
array = np.array([1, 2, 3, 4, 5])
array
```

```python
#adição
soma = np.add(array, 10)
soma
```

```python
#subtração
subtracao = np.subtract(array, 5)
subtracao
```

```python
#multiplicação
multiplicacao = np.multiply(array, 2)
multiplicacao
```

```python
#divisão
divisao = np.divide(array, 2)
divisao
```

```python
media = np.mean(array)
media
```

```python
desvio_padrao = np.std(array)
desvio_padrao
```

```python
comparação_maior_que_3 = np.greater(array, 3)
comparação_maior_que_3
```

```python
resultado = np.sqrt(array) + np.power(array, 2)
resultado
```

```python
arredondamento = np.round(array, 3)
arredondamento
```
